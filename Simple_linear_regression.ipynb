```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/exercise-1/1-studyhours.csv')
```

```python
plt.scatter(df['Study Hours'], df['Exam Score'])
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```

```python
# indepent and dependent features
X=df[['Study Hours']] 
y = df['Exam Score']
```

```python
type(X)
```

```python
type(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=15)
```

```python
#Standardize the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

#On train data we use "fit_transform" → learn mean & std, then scale the data.
#On test data we use "transform" → we use the same mean & std from train, without learning again.
#If we use "fit_transform" on test data it causes DATA LEAKAGE
#Data leakage happens when the model learns information from the test set during training. 
#This makes the model look better in tests, but it fails in real life. 
```

```python
plt.scatter(X_train, y_train)
plt.xlabel("X_train")
plt.ylabel("y_train")
plt.show()
```

```python
#Time to fit the model
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X_train,y_train)

print("Coefficient : ",regression.coef_)
print("Intercept : ",regression.intercept_)
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,regression.predict(X_train),"r")
plt.show()
```

```python
#prediction with test data
```

```python
y_pred_test = regression.predict(X_test)
y_test
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,regression.predict(X_test),"r")
plt.show()
```

```python
# Performance metrics

from sklearn.metrics import mean_squared_error, mean_absolute_error

mse=mean_squared_error(y_test,y_pred_test)
mae=mean_absolute_error(y_test,y_pred_test)
rmse = np.sqrt(mse)

print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred_test)
print(score)
```

```python
#adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
